```python
import numpy as np
import matplotlib.pyplot as plt
from Layers import *
from Models import *
np.random.seed(42)
```

```python
# データの作成
batch_size = 5
x_train = np.random.randn(batch_size, 3)
y_train = np.random.randn(batch_size, 2)
```

```python
model = Model(layers=[
    Linear(3, 9),
    Linear(9, 2)
])

epochs = 1000
lr = 0.01
losses = []

for epoch in range(epochs):
    # フォワードパス
    output = model.forward(x_train)
    
    # 損失の計算 (例としてMSEを使用)
    loss = np.mean((output - y_train) ** 2)
    losses.append(loss)
    
    # バックワードパス
    grad = 2 * (output - y_train) / y_train.size
    model.backward(grad, lr)
    
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss}")
```

```python
# 損失の可視化
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.show()
```
